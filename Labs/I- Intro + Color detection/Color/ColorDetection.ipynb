### Assignment I.1: Color detection

In this assignment, we will read color images in RGB color space and convert them into HSV color space.We will detect objects based on their colors in HSV space.

#### A.

Read the image file 'data/fruits.jpg'.

*Hint:* OpenCV package read image in BGR order, but still in RGB color space.

**Answer:**  

```python
fruits = cv2.imread('data/fruits.jpg',cv2.IMREAD_COLOR)
```

#### B.

Convert the color space into HSV space. Save this HSV array as **im_hsv**. Compare RGB image array and HSV image array.

*Hint:* Use the same function the one converts color image into gray.

**Answer:**  

What does 'H', 'S' and 'V' mean? What are the ranges of 'H', 'S' and 'V'?

```python
im_hsv = cv2.cvtColor(fruits, cv2.COLOR_BGR2HSV)
im_rgb = cv2.cvtColor(fruits, cv2.COLOR_BGR2RGB)
plt.imshow(im_rgb)
```

**Answer:**  

What is the range of 'H' of **im_hsv**? What color does it correspond to?

**Answer:**  

```python
print("Hue is in range from "+ str(np.amin(im_hsv[:,:,0])) +" to " +str(np.amax(im_hsv[:,:,0])))
print("Saturation is in range from "+ str(np.amin(im_hsv[:,:,1])) +" to " +str(np.amax(im_hsv[:,:,1])))
print("Value is in range from "+ str(np.amin(im_hsv[:,:,2])) +" to " +str(np.amax(im_hsv[:,:,2])))
```

#### C. 

Mask the green apple on the original image by using **im_hsv**, and show the masked image.

*Hint:*   
1. Set a range of green color.  
2. Threshold the HSV image by this range.(**cv2.inRange()**)  
3. Combine the mask and the original image.(**cv2.bitwise_and()**)

**Answer:**  

```python
print(im_hsv.shape)
from matplotlib.colors import hsv_to_rgb
import cv2
light_green = (165, 255, 255)
dark_green = (70, 70, 130)
im_hsv = cv2.cvtColor(fruits, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(im_hsv, (20, 90, 130), (40, 255,255))

print("Hue is in range from "+ str(np.amin(mask)) +" to " +str(np.amax(mask)))
print("Saturation is in range from "+ str(np.amin(mask)) +" to " +str(np.amax(mask)))
print("Value is in range from "+ str(np.amin(mask)) +" to " +str(np.amax(mask)))
    # Bitwise-AND mask and original image
result = cv2.bitwise_and(fruits,fruits, mask= mask)
plt.show()
plt.imshow(mask, cmap="gray")
plt.show()
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
```
