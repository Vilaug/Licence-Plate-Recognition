### Assignment I.2: Min/Max frequency in an image

In this assignment we will build our image. We will see the max and min frequencies that can be represented.

#### A.

Create an array of zeros of size 10\*10 and save it in a variable with the name **binary_image**.

*Hint:*  
1. Using **numpy** package  
2. dtype = np.uint8

**Answer:**  

```python
binary_image = np.zeros((10,10), dtype=np.uint8)
plt.imshow(binary_image)
f = np.fft.fft2(binary_image)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))
print(fshift)
plt.subplot(121),plt.imshow(binary_image, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

What is the frequency of the output image?

**Answer:**  

#### B.

Change the values of all the even columns of the **binary_image** matrix to ones. View the **binary_image** matrix using the command **plt.imshow**.

*Hint:*  
1. **binary_image** is colored image or gray_scale image
2. Python starts from 0.

**Answer:**  

```python
bin1 = binary_image
for i in range(5):
    bin1[:,i*2+1:i*2+2] = np.ones((10,1))
    
plt.imshow(bin1)
dft = cv2.dft(np.float32(bin1),flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))

plt.subplot(121),plt.imshow(bin1, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
binary_image = np.zeros((10,10), dtype=int)
```
